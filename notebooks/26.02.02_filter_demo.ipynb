```python
import poolparty as pp; pp.init()

def gc_content(seq: str) -> float:
    return (seq.count("G") + seq.count("C")) / len(seq)

print("=== All IUPAC sequences (4-mer with 2 variable positions) ===")
pool = pp.from_iupac("ACNN", mode="sequential")
pool.print_library()

print("\n=== Filter: keep only GC content >= 0.5 ===")
pool.filter(lambda s: gc_content(s) >= 0.5).print_library()

print("\n=== Filter: keep only GC content < 0.5 ===")
pool.filter(lambda s: gc_content(s) < 0.5).print_library()

print("\n=== Longer IUPAC sequence, filter for high GC ===")
pp.from_iupac("NNNN", mode="sequential").filter(lambda s: gc_content(s) >= 0.75).print_library()
```
