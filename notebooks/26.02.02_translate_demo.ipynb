```python
import poolparty as pp; pp.init()

print("Basic translation")
pp.from_seq("ATGGCTTAA").translate().print_library()

print("Translation without stop codon")
pp.from_seq("ATGGCTTAA").translate(include_stop=False).print_library()

print("Translation with ORF annotation")
pp.from_seq("NNATGGCTTAANN").annotate_orf("cds", extent=(2,11), frame=1).translate(region="cds").print_library()

print("Translation with codon styling")
pp.from_seq("ATGGCTTAA").annotate_orf("cds", frame=1, style_codons=["red","green","blue"]).translate(region="cds", preserve_codon_styles=True).print_library()

print("Translation with frame +2")
pp.from_seq("ATGGCTTAA").translate(frame=2).print_library()

print("Translation with frame +3")
pp.from_seq("ATGGCTTAA").translate(frame=3).print_library()
```
