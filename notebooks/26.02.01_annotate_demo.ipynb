```python
import poolparty as pp; pp.init()

print("annotate_region() inserts XML tags to define a region")
pp.from_seq("NNNNATGCCCGGGTAANNN").annotate_region("orf", extent=(4, 16)).print_library()

print("\nannotate_orf() defines an ORF with persistent frame metadata")
pp.init()
pp.from_seq("NNNNATGCCCGGGTAANNN").annotate_orf("orf", extent=(4, 16), frame=1).print_library()

print("\nannotate_orf() with style_codons applies codon coloring")
pp.init()
pp.from_seq("NNNNATGCCCGGGTAANNN").annotate_orf("orf", extent=(4, 16), frame=1, style_codons=["red", "blue"]).print_library()

print("\nmutagenize_orf() reads frame automatically from OrfRegion")
pp.init()
pp.from_seq("<orf>ATGCCCGGGTAA</orf>").annotate_orf("orf", frame=2).mutagenize_orf(region="orf", num_mutations=1).print_library()

print("\nstylize_orf() also reads frame automatically from OrfRegion")
pp.init()
pp.from_seq("<orf>ATGCCCGGGTAA</orf>").annotate_orf("orf", frame=1).stylize_orf(region="orf", style_codons=["cyan", "magenta"]).print_library()

print("\nannotate_orf() upgrades a plain Region to OrfRegion")
pp.init()
pp.from_seq("<target>ATGAAACCC</target>").annotate_orf("target", frame=1).mutagenize_orf(region="target", num_mutations=1).print_library()
```
