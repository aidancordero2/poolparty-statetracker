```python
import poolparty as pp; pp.init()

# Blink style demo - Note: blink rendering depends on terminal support
# 'blink' and 'blinking' are aliases for the ANSI blink code (5)

# Basic blink
pool = pp.from_seq("ACGTACGT", style="blink").named("basic_blink")
pool.print_library()

# Blink with color
pool = pp.from_seq("ACGTACGT", style="blink red").named("blink_red")
pool.print_library()

# Blink with multiple modifiers
pool = pp.from_seq("ACGTACGT", style="blink bold cyan").named("blink_bold_cyan")
pool.print_library()

# 'blinking' alias works the same
pool = pp.from_seq("ACGTACGT", style="blinking magenta").named("blinking_alias")
pool.print_library()

# Blink only on mutated positions
pool = pp.from_seq("ACGTACGT").mutagenize(num_mutations=2, style="blink red lower", mode="sequential").named("mut_blink")
pool.print_library(num_seqs=5)

# Invert style demo - swaps foreground and background colors
# 'invert' and 'reverse' are aliases for the ANSI reverse video code (7)

# Basic invert
pool = pp.from_seq("ACGTACGT", style="invert").named("basic_invert")
pool.print_library()

# Invert with color
pool = pp.from_seq("ACGTACGT", style="invert red").named("invert_red")
pool.print_library()

# 'reverse' alias works the same
pool = pp.from_seq("ACGTACGT", style="reverse cyan").named("reverse_alias")
pool.print_library()

# Invert on mutated positions
pool = pp.from_seq("ACGTACGT").mutagenize(num_mutations=2, style="invert lower", mode="sequential").named("mut_invert")
pool.print_library(num_seqs=5)
```
