```python
import poolparty as pp; pp.init()

# Create a pool with multiple sequences (mode='sequential' gives one state per sequence)
pool = pp.from_seqs(['ACGTACGTACGT', 'GGGAAACCCGGG', 'TTTACGTTTAAA'], mode='sequential').named('seqs')

# Create highlighters for different patterns
hl_acgt = pp.Highlighter(r'ACGT', 'red')        # Highlight "ACGT" in red
hl_aaa = pp.Highlighter(r'AAA', 'underline')   # Highlight "AAA" with underline
hl_ggg = pp.Highlighter(r'GGG', 'cyan')        # Highlight "GGG" in cyan

# Print with multiple overlapping highlights
pool.print_library(highlights=[hl_acgt, hl_aaa, hl_ggg])
```
