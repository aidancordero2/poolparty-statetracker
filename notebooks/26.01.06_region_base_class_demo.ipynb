```python
import poolparty as pp; pp.init()

# Demo: region and remove_marker parameters work consistently across operation types
# These parameters are now handled by the Operation base class

bg = pp.from_seq('AAAA<region>acAC</region>TTTT')
print("Source sequence with marker:"); bg.print_library()

# 1. FixedOp (via upper) - with and without marker removal
print("1. upper() with region='region':")
pp.upper(bg, region='region').print_library()
print("   upper() with region='region', remove_marker=False:")
pp.upper(bg, region='region', remove_marker=False).print_library()
print("   upper() with region='region', remove_marker=True:")
pp.upper(bg, region='region', remove_marker=True).print_library()

# 2. shuffle_seq with region and remove_marker
print("2. shuffle_seq() with region='region', remove_marker=True:")
pp.shuffle_seq(bg, region='region', remove_marker=True).print_library(seed=42)
print("   shuffle_seq() with region='region', remove_marker=False:")
pp.shuffle_seq(bg, region='region', remove_marker=False).print_library(seed=42)

# 3. mutagenize with region and remove_marker
print("3. mutagenize() with region='region', remove_marker=True:")
pp.mutagenize(bg, num_mutations=1, region='region', remove_marker=True).print_library(seed=42)
print("   mutagenize() with region='region', remove_marker=False:")
pp.mutagenize(bg, num_mutations=1, region='region', remove_marker=False).print_library(seed=42)

# 4. All operations also support [start, stop] interval syntax for region
print("4. Using [start, stop] interval syntax (positions 4-8):")
pp.upper('aaaacccctttt', region=[4, 8]).print_library()
```
