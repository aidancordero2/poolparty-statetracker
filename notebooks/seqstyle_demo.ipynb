```python
import poolparty as pp
pp.init()

from poolparty.utils.style_utils import SeqStyle
import numpy as np

# Create a styled sequence
style_list = [('red', np.array([0, 1, 2])), ('blue', np.array([5, 6, 7]))]
s = SeqStyle.from_style_list(style_list, length=10)
print(f"Original: {s}")  # SeqStyle(2 styles, length=10)
print(f"style_list: {s.style_list}")

# Slicing extracts and 0-indexes
region = s[2:8]
print(f"\ns[2:8]: {region}")
print(f"style_list: {region.style_list}")

# Reversal for rc operations
rev = s[:5].reversed()
print(f"\ns[:5].reversed(): {rev}")
print(f"style_list: {rev.style_list}")

# Concatenation
combined = s[:3] + SeqStyle.empty(2) + s[7:]
print(f"\nCombined s[:3] + empty(2) + s[7:]: {combined}")
print(f"style_list: {combined.style_list}")

# Add a style (returns new SeqStyle)
with_extra = s.add_style('green', np.array([8, 9]))
print(f"\nWith extra style: {with_extra}")
print(f"style_list: {with_extra.style_list}")

# Split at multiple breakpoints
parts = s.split([3, 7])  # Returns 3 parts
print(f"\nSplit into {len(parts)} parts: {[str(p) for p in parts]}")

# Bool check
print(f"\nbool(s): {bool(s)}")  # True
print(f"bool(SeqStyle.empty(10)): {bool(SeqStyle.empty(10))}")  # False

# Applying styles to a sequence
seq = "ACGTACGTAC"
styled_seq_style = SeqStyle.from_style_list([('red', np.array([0, 1, 2])), ('blue', np.array([7, 8, 9]))], length=10)
print(f"\nApplying styles to sequence:")
print(styled_seq_style.apply(seq))

# Validation
print(f"\nValidation:")
try:
    bad_style = SeqStyle.from_style_list([('red', np.array([0, 10]))], length=10)
    bad_style.validate()
except ValueError as e:
    print(f"Caught validation error: {e}")
```
