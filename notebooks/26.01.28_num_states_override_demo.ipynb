```python
import poolparty as pp; pp.init()
from poolparty.base_ops.from_seqs import from_seqs
from poolparty.base_ops.mutagenize import mutagenize
from poolparty.base_ops.from_iupac import from_iupac

print("="*60)
print("DEMO: num_states override in sequential mode")
print("="*60)

# Example 1: from_seqs with clipping (num_states < natural)
print("\n1. from_seqs: Clipping (num_states=2, natural=4)")
print("-"*50)
pool = from_seqs(['A', 'B', 'C', 'D'], mode='sequential', num_states=2).named('clipped')
print(f"   num_states={pool.operation.num_states}, natural_num_states={pool.operation.natural_num_states}")
pool.print_library(num_seqs=6)

# Example 2: from_seqs with cycling (num_states > natural)
print("\n2. from_seqs: Cycling (num_states=8, natural=2)")
print("-"*50)
pool = from_seqs(['A', 'B'], mode='sequential', num_states=8).named('cycled')
print(f"   num_states={pool.operation.num_states}, natural_num_states={pool.operation.natural_num_states}")
pool.print_library(num_seqs=10)

# Example 3: mutagenize with clipping
print("\n3. mutagenize: Clipping (num_states=4, natural=6)")
print("-"*50)
pool = mutagenize('AC', num_mutations=1, mode='sequential', num_states=4).named('mut_clipped')
print(f"   num_states={pool.operation.num_states}, natural_num_states={pool.operation.natural_num_states}")
pool.print_library(num_seqs=8)

# Example 4: mutagenize with cycling
print("\n4. mutagenize: Cycling (num_states=12, natural=6)")
print("-"*50)
pool = mutagenize('AC', num_mutations=1, mode='sequential', num_states=12).named('mut_cycled')
print(f"   num_states={pool.operation.num_states}, natural_num_states={pool.operation.natural_num_states}")
pool.print_library(num_seqs=14)

# Example 5: from_iupac with clipping
print("\n5. from_iupac: Clipping 'RY' (num_states=2, natural=4)")
print("-"*50)
pool = from_iupac('RY', mode='sequential', num_states=2).named('iupac_clipped')
print(f"   num_states={pool.operation.num_states}, natural_num_states={pool.operation.natural_num_states}")
pool.print_library(num_seqs=6)

# Example 6: Practical use case - generating exactly 100 sequences from a 30-state pool
print("\n6. Practical use case: Generate exactly 100 sequences")
print("-"*50)
pool = mutagenize('ACGTACGTAC', num_mutations=1, mode='sequential', num_states=100).named('library')
print(f"   Natural states for 10bp + 1 mutation = 10*3 = 30")
print(f"   num_states={pool.operation.num_states}, natural_num_states={pool.operation.natural_num_states}")
print(f"   Generates 100 sequences by cycling through 30 states 3+ times")
pool.print_library(num_seqs=10)  # Just show first 10
print("   ... (90 more sequences)")

print("\n" + "="*60)
print("All examples demonstrate num_states override working correctly!")
print("="*60)
```
