```python
import poolparty as pp; pp.init()

print("Basic background colors")
pp.from_seq("ACGTACGT", style="on_red").named("on_red").print_library()
pp.from_seq("ACGTACGT", style="on_blue").named("on_blue").print_library()
pp.from_seq("ACGTACGT", style="on_green").named("on_green").print_library()
pp.from_seq("ACGTACGT", style="on_yellow").named("on_yellow").print_library()
pp.from_seq("ACGTACGT", style="on_white").named("on_white").print_library()
pp.from_seq("ACGTACGT", style="on_black").named("on_black").print_library()

print("CSS background colors")
pp.from_seq("ACGTACGT", style="on_coral").named("on_coral").print_library()
pp.from_seq("ACGTACGT", style="on_gold").named("on_gold").print_library()
pp.from_seq("ACGTACGT", style="on_darkviolet").named("on_darkviolet").print_library()

print("Hex background colors")
pp.from_seq("ACGTACGT", style="on_#ff7f50").named("on_#ff7f50").print_library()
pp.from_seq("ACGTACGT", style="on_#00ff00").named("on_#00ff00").print_library()

print("Foreground white and black")
pp.from_seq("ACGTACGT", style="white").named("white").print_library()
pp.from_seq("ACGTACGT", style="black").named("black").print_library()

print("Combined foreground and background")
pp.from_seq("ACGTACGT", style="white on_blue").named("white_on_blue").print_library()
pp.from_seq("ACGTACGT", style="black on_yellow").named("black_on_yellow").print_library()
pp.from_seq("ACGTACGT", style="red on_white").named("red_on_white").print_library()

print("Combined with modifiers")
pp.from_seq("ACGTACGT", style="white on_blue bold").named("white_on_blue_bold").print_library()
pp.from_seq("ACGTACGT", style="black on_yellow underline").named("black_on_yellow_underline").print_library()

print("Background on mutations")
pp.from_seq("ACGTACGT").mutagenize(num_mutations=2, style="lower on_red", mode="sequential").named("mut_on_red").print_library(num_seqs=5)
```
