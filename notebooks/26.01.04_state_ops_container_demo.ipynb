```python
import poolparty as pp
pp.init()
```

```python
# Create a pool with 5 states
pool = pp.from_seqs(['A', 'B', 'C', 'D', 'E'], mode='sequential').print_library()
```

```python
# sample_states(): sample 3 states with a seed for reproducibility
pool.sample_states(num_states=3, seed=42).print_library()
```

```python
# sample_states(): explicit state indices
pool.sample_states(sampled_states=[0, 2, 4]).print_library()
```

```python
# shuffle_states(): randomly permute the states
pool.shuffle_states(seed=42).print_library()
```

```python
# shuffle_states(): with explicit permutation (reverse order)
pool.shuffle_states(permutation=[4, 3, 2, 1, 0]).print_library()
```

```python
# slice_states(): select a range of states
pool.slice_states(slice(1, 4)).print_library()
```

```python
# slice_states(): select a single state by index
single = pool.slice_states(2).print_library()
```
