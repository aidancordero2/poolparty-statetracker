```python
import poolparty as pp; pp.init()

orf_seq = "ATGGCTAGCTACGAATGCAAGTACCCGTAA"
codon_style = ["gold", "cyan", "magenta"]
num_seqs = 5

for frame in [1, 2, 3, -1, -2, -3]:
    print(f"=== frame={frame} ===")
    orf = pp.from_seq(orf_seq).stylize_orf(style_codons=codon_style, frame=frame).named(f"frame{frame}").print_library()
    mut_seqential = orf.mutagenize_orf(num_mutations=1, mode="sequential", style="red bold", frame=frame)\
        .named(f"frame{frame}_sequential").print_library(num_seqs=num_seqs)
    mut_random = orf.mutagenize_orf(num_mutations=1, mode="random", style="red bold", frame=frame)\
        .named(f"frame{frame}_random").print_library(num_seqs=num_seqs)
```
