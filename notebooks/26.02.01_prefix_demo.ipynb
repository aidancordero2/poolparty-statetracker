```python
import poolparty as pp; pp.init()

print("=== Fixed operation prefix adds constant label ===")
pp.from_seq("ACGT", prefix="bg").upper(prefix="upper").print_library()

print("\n=== Combined fixed and variable ops ===")
pp.from_seq("ACGTACGT", prefix="bg").mutagenize(num_mutations=1, mode="sequential", prefix="mut").upper(prefix="upper").print_library()

print("\n=== Filter with prefix ===")
pp.from_seqs(["AAAA", "CCCC", "GGGG"], mode="sequential", prefix="seq").filter(lambda s: s.startswith("A"), prefix="filtered").print_library()

print("\n=== Region operations with prefix ===")
pp.from_seq("ACGTACGT", prefix="bg").insert_tags("region", start=2, stop=6, prefix="tagged").print_library()

print("\n=== Composed operations pass prefix to underlying op ===")
pp.from_seq("ACGTACGT", prefix="bg").annotate_region("orf", extent=(2, 6), prefix="orf_tag").print_library()
```
