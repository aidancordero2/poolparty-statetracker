```python
# Demonstration of per-sequence inline styling
import poolparty as pp
pp.init()

# Create a template with a marked region
pool = pp.from_seq('TCCCGACT<cre>GGAAAGCGGGCAGTGAGCACACAGG</cre>ATTACGGATGCA')\
    .named('template')\
    .stylize('cre', style='salmon')

# Mutagenize with changes_style - mutations will be styled
# The cyan+bold style is applied to each mutated position
mutated = pool.mutagenize('cre', 
                          mutation_rate=0.15,
                          mode='random',
                          num_states=5,
                          changes_style='cyan bold lower',  # Style mutations in cyan+bold
                          seq_name_prefix='mut_').named('mutated')

# print_library() applies the inline styles to each sequence
# The cyan highlights appear on the lowercase letters (mutations)
mutated.print_library(seed=42)
```

```python
# You can also combine inline styles with global highlights
pp.clear_highlights()
pp.add_highlight(which='tags', style='gray')  # Style XML tags gray
pp.add_highlight(region='cre', which='upper', style='red')  # Style uppercase in region red

# Now mutations appear cyan+bold on top of the red background
mutated.print_library(seed=42)
```

```python
# Demo of 'upper' and 'lower' case transforms in inline styles
# These transform the character case at styled positions
pp.clear_highlights()

# Create a simple uppercase sequence
pool = pp.from_seq('AAAA<test>GGGGCCCC</test>TTTT').named('template')

# 'lower cyan bold' - convert mutations to lowercase AND style cyan+bold
mutated = pool.mutagenize('test',
                          mutation_rate=0.5,
                          mode='random',
                          num_states=3,
                          changes_style='lower cyan bold',  # lowercase + cyan + bold
                          seq_name_prefix='mut_').named('mutated_lower')
mutated.print_library(seed=42)

# 'upper red' on a lowercase template - convert mutations to uppercase AND style red
pool2 = pp.from_seq('aaaa<test>ggggcccc</test>tttt').named('template2')
mutated2 = pool2.mutagenize('test',
                            mutation_rate=0.5,
                            mode='random',
                            num_states=3,
                            changes_style='upper red',  # uppercase + red
                            seq_name_prefix='mut_').named('mutated_upper')
mutated2.print_library(seed=42)
```
