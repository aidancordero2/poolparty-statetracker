```python
import poolparty as pp
```

```python
# Sequence names propagate from parent to child by default
pp.init()
parent = pp.from_seqs(['ACGT', 'TGCA'], seq_names=['seq_a', 'seq_b'], mode='sequential')
mutants = pp.mutagenize(parent, num_mutations=1, mode='sequential').named('mut')
mutants.generate_library(num_cycles=1)
```

```python
# clear_seq_names() blocks name propagation, returns None for all names
pp.init()
parent = pp.from_seqs(['ACGT', 'TGCA'], seq_names=['seq_a', 'seq_b'], mode='sequential')
mutants = pp.mutagenize(parent, num_mutations=1, mode='sequential').named('mut').clear_seq_names()
mutants.generate_library(num_cycles=1)
```

```python
# Also works with explicit seq_names on from_seqs
pp.init()
pool = pp.from_seqs(
        ['AAA', 'TTT', 'GGG'],
        seq_names=['alpha', 'beta', 'gamma'],
        mode='sequential',
    ).named('pool').clear_seq_names()

pool.generate_library(num_cycles=1)
```

```python
# Works with stacked pools - blocks names from all parents
with pp.Party() as party:
    a = pp.from_seqs(['AAA'], seq_names=['name_a'], mode='sequential')
    b = pp.from_seqs(['TTT'], seq_names=['name_b'], mode='sequential')
    stacked = (a + b).named('stacked').clear_seq_names()

stacked.generate_library(num_cycles=1)
```

```python
# Method chaining: clear_seq_names() returns self
with pp.Party() as party:
    pool = pp.from_seqs(['ACGT'], seq_names=['test'], mode='sequential').named('pool').clear_seq_names()
    print(f"pool.name = {pool.name}")
    print(f"pool.operation._block_seq_names = {pool.operation._block_seq_names}")
```
