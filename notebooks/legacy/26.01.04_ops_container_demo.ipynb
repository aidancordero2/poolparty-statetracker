```python
import poolparty as pp
```

```python
# pool.ops.mutagenize() applies mutagenize() to a marked region
with pp.Party():
    bg = pp.from_seq('FLANK<target>ACGT</target>FLANK')
    result = bg.ops.mutagenize('target', num_mutations=1, mark_changes=True, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# pool.ops.deletion_scan() applies deletion_scan() to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGTACGT</region>TTT')
    result = bg.ops.deletion_scan('region', deletion_length=2, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# pool.ops.insertion_scan() applies insertion_scan() to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGT</region>TTT')
    result = bg.ops.insertion_scan('region', 'XX', mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# pool.ops.replacement_scan() applies replacement_scan() to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGTACGT</region>TTT')
    result = bg.ops.replacement_scan('region', 'NN', mark_changes=True, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# pool.ops.mutagenize_scan() applies mutagenize_scan() to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGTACGT</region>TTT')
    result = bg.ops.mutagenize_scan('region', mutagenize_length=4, num_mutations=1, mode='sequential')
    df = result.generate_library(num_cycles=1, seed=42)
df
```

```python
# pool.ops.shuffle_scan() applies shuffle_scan() to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGTACGT</region>TTT')
    result = bg.ops.shuffle_scan('region', shuffle_length=4, mark_changes=True, mode='sequential')
    df = result.generate_library(num_cycles=1, seed=42)
df
```

```python
# pool.ops.shuffle_seq() shuffles the content of a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGT</region>TTT')
    result = bg.ops.shuffle_seq('region')
    df = result.generate_library(num_cycles=3, seed=42)
df
```

```python
# pool.ops.from_iupac_motif() replaces marker with IUPAC-generated sequences
with pp.Party():
    bg = pp.from_seq('AAA<insert/>TTT')
    result = bg.ops.from_iupac_motif('insert', 'RN', mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# pool.ops.get_kmers() replaces marker with k-mer sequences
with pp.Party():
    bg = pp.from_seq('AAA<insert/>TTT')
    result = bg.ops.get_kmers('insert', length=2, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# pool.ops.from_prob_motif() replaces marker with probability-sampled sequences
import pandas as pd
prob_df = pd.DataFrame({'A': [0.5, 0.0], 'C': [0.5, 0.0], 'G': [0.0, 0.5], 'T': [0.0, 0.5]})
with pp.Party():
    bg = pp.from_seq('AAA<insert/>TTT')
    result = bg.ops.from_prob_motif('insert', prob_df, mode='random')
    df = result.generate_library(num_cycles=3, seed=42)
df
```

```python
# pool.ops.apply_at_marker() applies a custom transform to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<target>ACGT</target>TTT')
    result = bg.ops.apply_at_marker('target', lambda p: pp.mutagenize(p, num_mutations=1))
    df = result.generate_library(num_cycles=1, seed=42)
df
```
