```python
import poolparty as pp; pp.init()

# Basic deletion scan with default marker
with pp.Party():
    pool = pp.from_seq('ACGTACGTACGT')
    result = pool.deletion_scan(deletion_length=3, mode='sequential')
    result.print_library()

# Deletion without marker (removes segment)
with pp.Party():
    pool = pp.from_seq('ACGTACGTACGT')
    result = pool.deletion_scan(deletion_length=3, deletion_marker=None, mode='sequential')
    result.print_library()

# Deletion with custom marker and style
with pp.Party():
    pool = pp.from_seq('ACGTACGTACGT')
    result = pool.deletion_scan(deletion_length=2, deletion_marker='*', style='red bold', mode='sequential')
    result.print_library()

# Deletion at specific positions
with pp.Party():
    pool = pp.from_seq('ACGTACGTACGT')
    result = pool.deletion_scan(deletion_length=4, positions=[0, 4, 8], mode='sequential')
    result.print_library()
```
