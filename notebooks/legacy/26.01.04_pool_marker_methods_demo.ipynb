```python
import poolparty as pp
```

```python
# Pool.mutagenize() applies mutagenize() to a marked region
with pp.Party():
    bg = pp.from_seq('FLANK<target>ACGT</target>FLANK')
    result = bg.mutagenize('target', num_mutations=1, mark_changes=True, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# Pool.deletion_scan() applies deletion_scan() to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGTACGT</region>TTT')
    result = bg.deletion_scan('region', deletion_length=2, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# Pool.insertion_scan() applies insertion_scan() to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGT</region>TTT')
    result = bg.insertion_scan('region', 'XX', mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# Pool.replacement_scan() applies replacement_scan() to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGTACGT</region>TTT')
    result = bg.replacement_scan('region', 'NN', mark_changes=True, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# With remove_marker=False, marker tags are preserved around the transformed content
with pp.Party():
    bg = pp.from_seq('FLANK<target>ACGT</target>FLANK')
    result = bg.mutagenize('target', remove_marker=False, num_mutations=1, mark_changes=True, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# Pool.mutagenize_scan() applies mutagenize_scan() to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGTACGT</region>TTT')
    result = bg.mutagenize_scan('region', mutagenize_length=4, num_mutations=1, mode='sequential')
    df = result.generate_library(num_cycles=1, seed=42)
df
```

```python
# Pool.shuffle_scan() applies shuffle_scan() to a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGTACGT</region>TTT')
    result = bg.shuffle_scan('region', shuffle_length=4, mark_changes=True, mode='sequential')
    df = result.generate_library(num_cycles=1, seed=42)
df
```

```python
# Pool.shuffle_seq() shuffles the content of a marked region
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGT</region>TTT')
    result = bg.shuffle_seq('region')
    df = result.generate_library(num_cycles=3, seed=42)
df
```

```python
# Pool.from_iupac_motif() replaces marker with IUPAC-generated sequences
with pp.Party():
    bg = pp.from_seq('AAA<insert/>TTT')
    result = bg.from_iupac_motif('insert', 'RN', mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# Pool.get_kmers() replaces marker with k-mer sequences
with pp.Party():
    bg = pp.from_seq('AAA<insert/>TTT')
    result = bg.get_kmers('insert', length=2, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# With remove_marker=False on a zero-length marker, tags wrap the inserted content
with pp.Party():
    bg = pp.from_seq('AAA<insert/>TTT')
    result = bg.get_kmers('insert', length=2, remove_marker=False, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```
