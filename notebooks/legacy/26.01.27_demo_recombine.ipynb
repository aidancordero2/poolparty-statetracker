```python
import poolparty as pp; pp.init()

# Demo: Recombine Operation - Simulate evolutionary recombination

# Create two source pools with distinct sequences
pool_a = pp.from_seq('AAAAAAAA').stylize(style='red')
pool_b = pp.from_seq('TTTTTTTT').stylize(style='blue')

# Basic recombination with 1 breakpoint
print("=" * 60)
print("Basic recombination with 1 breakpoint")
print("=" * 60)
pp.recombine(
    sources=[pool_a, pool_b],
    num_breakpoints=1,
    mode='sequential'
).print_library()

# Multiple breakpoints
print("\n" + "=" * 60)
print("Recombination with 2 breakpoints")
print("=" * 60)
pp.recombine(
    sources=[pool_a, pool_b],
    num_breakpoints=2,
    mode='sequential'
).print_library()

# Random mode with multiple source pools
print("\n" + "=" * 60)
print("Random mode with 3 source pools")
print("=" * 60)
pool_c = pp.from_seq('GGGGGGGG')
pp.recombine(
    sources=[pool_a, pool_b, pool_c],
    num_breakpoints=2,
    mode='random',
    num_states=10
).print_library(seed=42)

# Style application to segments
print("\n" + "=" * 60)
print("Recombination with segment styling")
print("=" * 60)
pp.recombine(
    sources=[pool_a, pool_b],
    num_breakpoints=1,
    positions=[3],
    mode='fixed',
    styles=['red', 'blue']  # Style segment 0 red, segment 1 blue
).print_library()

# Region-based recombination
print("\n" + "=" * 60)
print("Region-based recombination")
print("=" * 60)
background = pp.from_seq('NNNNNNNNNNNN').insert_tags('insert', start=4, stop=8)
background.recombine(
    region='insert',
    sources=[pool_a, pool_b],
    num_breakpoints=1,
    positions=[1],
    mode='fixed'
).print_library()

# Chained operations with style inheritance
print("\n" + "=" * 60)
print("Chained operations with style inheritance")
print("=" * 60)
styled_a = pp.from_seq('ACGTACGT').mutagenize(num_mutations=1, style='green', mode='random', num_states=1)
styled_b = pp.from_seq('TGCATGCA').mutagenize(num_mutations=1, style='yellow', mode='random', num_states=1)
pp.recombine(
    sources=[styled_a, styled_b],
    num_breakpoints=1,
    positions=[3],
    mode='fixed',
    styles=['bold', '']  # Bold on segment 0, no additional style on segment 1
).print_library(seed=42)
```
