```python
import poolparty as pp
import statecounter as sc
pp.init()
```

```python
# Create two pools via ordered_product to show default behavior
A = pp.from_seqs(['a1', 'a2'], name='A')
B = pp.from_seqs(['b1', 'b2', 'b3'], name='B')
C = pp.join([A, B], name='AB')
C.generate_library(num_cycles=1)
```

```python
# Default: 'first_counter_fastest' - lower ID counters (A) cycle fastest
sc.get_product_order_mode()
```

```python
# Switch to 'first_counter_slowest' mode using pp.set_default
pp.set_default('iter_order', 'first_counter_slowest')
sc.get_product_order_mode()
```

```python
# Now A (first created) will cycle slowest, B cycles fastest
pp.init()
A = pp.from_seqs(['a1', 'a2'], name='A')
B = pp.from_seqs(['b1', 'b2', 'b3'], name='B')
C = pp.join([A, B], name='AB')
C.generate_library(num_cycles=1)
```

```python
# Reset to default behavior
pp.set_default('iter_order', 'first_counter_fastest')
sc.get_product_order_mode()
```
