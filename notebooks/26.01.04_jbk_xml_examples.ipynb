```python
import poolparty as pp
pp.init()

template = 'TCCCGACT<mut>GGAAAGCGGGCAG<misc/>TGAGCACACAGGAAA</mut>ATTACGG<bc/>AGATCGGA'

pool1 = pp.from_seq(template)\
           .named('pool1')\
           .print_library()

pool2 = pool1.mutagenize('mut', mutation_rate=0.25, mode='hybrid', num_hybrid_states=3)\
           .repeat_states(5)\
           .insert_kmers('bc', length=5, case='lower')\
           .remove_marker('misc')\
           .named('pool2')\
           .print_library()

pool3 = pool2.clear_markers()\
           .upper()\
           .named('pool3')\
           .print_library()
```
