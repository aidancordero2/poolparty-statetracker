```python
import poolparty as pp
pp.reset_default_party()

marked_seq = 'TCCCGACT.<mut>GGAAAGCGGGCAGTGAGCACACAGGAAA</mut>.ATTACGG.<bc/>.AGATCGGA'
pool = pp.from_seq(marked_seq)\
        .ops.mutagenize('mut',mutation_rate=0.25,remove_marker=False,mark_changes=True)\
        .ops.get_kmers('bc',length=5,remove_marker=False)\
        .ops.clear_markers()
pool.generate_library(num_seqs=10, seqs_only=True)
```
