```python
import poolparty as pp
pp.init()

template = pp.from_seq('TCCCGACT<cre>GGAAAGCGGGCAGTGAGCACACAGGAAA</cre>ATTACGG<bc/>AGATCGGA')

# Use seq_name_prefix directly in repeat() instead of name_seqs()
wt_pool = (
    template
    .named('wt_pool')
    .repeat_states(1, seq_name_prefix='wt.v')
    .print_library()    
)

mutagenized_pool = (
    template
    .mutagenize('cre', mutation_rate=0.25, mode='hybrid', num_hybrid_states=5, seq_name_prefix='mut_')
    .named('mutagenized_pool')
    .repeat_states(3, seq_name_prefix='v', op_iter_order=-1)
    .print_library()
)


final_pool = (
    pp.stack([
        wt_pool,
        mutagenized_pool,
        ])
    .named('barcoded_pool')
    .insert_kmers('bc', length=5, case='lower')
    .clear_nonmolecular_chars()
    .print_library()
)
```
