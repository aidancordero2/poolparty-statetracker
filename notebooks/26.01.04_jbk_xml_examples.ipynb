```python
import poolparty as pp
pp.init()

template = pp.from_seq('TCCCGACT<cre>GGAAAGCGGGCAGTGAGCACACAGGAAA</cre>ATTACGG<bc/>AGATCGGA')

wt_pool = (
    template
    .named('wt_pool')
    .repeat_states(10)
    .name_seqs(prefix='wt.v')
    .print_library()    
)

mutagenized_pool = (
    template
    .mutagenize('cre', mutation_rate=0.25, mode='hybrid', num_hybrid_states=5)
    .named('mutagenized_pool')
    .name_seqs(prefix='mut_')
    .print_library()
)

deletion_pool = (
    template
    .deletion_scan('cre', deletion_length=5, positions=slice(None,None,5), mode='sequential')
    .named('deletion_pool')
    .name_seqs(prefix='del_')
    .print_library()
)

final_pool = (
    pp.stack([
        wt_pool,
        mutagenized_pool,
        deletion_pool,
        ])
    .named('barcoded_pool')
    .insert_kmers('bc', length=5, case='lower')
    .clear_nonmolecular_chars()
    .print_library()
)


```

```python
final_pool = (
    pp.stack([
        wt_pool.repeat_states(3).name_seqs(prefix='v'),
        mutagenized_pool.repeat_states(3).name_seqs(prefix='v'),
        deletion_pool.repeat_states(3).name_seqs(prefix='v'),
        ])
    .named('barcoded_pool2')
    .insert_kmers('bc', length=5, case='lower')
    .clear_nonmolecular_chars()
    .print_library()
)
```
