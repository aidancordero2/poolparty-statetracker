```python
import poolparty as pp
```

```python
# upper() - convert to uppercase (preserves markers)
with pp.Party():
    pool = pp.from_seq('acgt<region strand="-">tt</region>')
    result = pp.upper(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# lower() - convert to lowercase (preserves markers)
with pp.Party():
    pool = pp.from_seq('ACGT<region strand="-">TT</region>')
    result = pp.reverse_complement(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# swap_case() - swap uppercase/lowercase (preserves markers)
with pp.Party():
    pool = pp.from_seq('ACgt<m>XX</m>')
    result = pp.swap_case(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# reverse_complement()
with pp.Party():
    pool = pp.from_seq('ACGT')
    result = pp.reverse_complement(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# clear_ignore_chars() - remove gaps/spaces, keep markers
with pp.Party():
    pool = pp.from_seq('A-C-G-T<m>XX</m>')
    result = pp.clear_ignore_chars(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# clear_nonmolecular_chars() - remove gaps AND markers, keep only alphabet chars
with pp.Party():
    pool = pp.from_seq('A-C-G-T<m>XX</m>')
    result = pp.clear_nonmolecular_chars(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# OpsContainer convenience methods (pool.ops.*)
with pp.Party():
    pool = pp.from_seq('ACgt')
    rc = pool.ops.reverse_complement()
    sc = pool.ops.swap_case()
    up = pool.ops.upper()
    lo = pool.ops.lower()
print('reverse_complement:', rc.generate_library(num_cycles=1, seqs_only=True))
print('swap_case:', sc.generate_library(num_cycles=1, seqs_only=True))
print('upper:', up.generate_library(num_cycles=1, seqs_only=True))
print('lower:', lo.generate_library(num_cycles=1, seqs_only=True))
```

```python
# clear_ignore_chars vs clear_nonmolecular_chars via OpsContainer
with pp.Party():
    pool = pp.from_seq('A-C-G-T<m>X</m>')
    ci = pool.ops.clear_ignore_chars()
    cn = pool.ops.clear_nonmolecular_chars()
print('clear_ignore_chars:', ci.generate_library(num_cycles=1, seqs_only=True))
print('clear_nonmolecular:', cn.generate_library(num_cycles=1, seqs_only=True))
```
