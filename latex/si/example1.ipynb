```python
import poolparty as pp

# Define sequences (lowercase wild-type so changes appear uppercase)
wt_seq = 'ATCGATCGATCGATCGATCG'.lower()          # 20 nt wild-type
adapter_5_seq = 'GAATTC'                          # 6 nt 5' adapter
adapter_3_iupac = 'GGATCC.NNNNNNNNNN.CTGCAG'      # 6 + 10N barcode + 6

with pp.Party() as party:
    # Create pools
    mut_pool = pp.mutagenize(
        wt_seq, 
        mutation_rate=0.10, 
        mark_changes=True, 
        op_name='mut')
    adapter_3_pool = pp.from_iupac_motif(
        adapter_3_iupac, 
        mark_changes=True, 
        op_name='iupac')

    # Join into final oligo
    oligo_pool_1 = pp.join(
        [adapter_5_seq, mut_pool, adapter_3_pool], 
        spacer_str='.',
        op_name='join')

# Sample 5 sequences
df = oligo_pool_1.generate_seqs(num_seqs=5, seed=42)
print(df['seq'].tolist())
```
