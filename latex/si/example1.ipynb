```python
import poolparty as pp

# Define sequences (lowercase wild-type so changes appear uppercase)
wt_seq = 'ATCGATCGAT'                  # 10 nt wild-type
adapter_5_seq = 'GAA'                  # 3 nt 5' adapter
adapter_3_iupac = 'GGA.NNNNN.CAG'.lower()      # 3 + 5N barcode + 3

# Create pool of randomly mutagenized wt_seq sequences
mut_pool = pp.mutagenize(
    wt_seq, 
    mutation_rate=0.10, 
    mark_changes=True, 
    op_name='op1')

# Create pool of 3' adaptors with random barcodes
adapter_3_pool = pp.from_iupac_motif(
    adapter_3_iupac, 
    mark_changes=False, 
    op_name='op2')

# Join into final oligo pool
oligo_pool = pp.join(
    [adapter_5_seq, mut_pool, adapter_3_pool], 
    spacer_str=' ',
    op_name='op3')

# Sample 5 sequences
df = oligo_pool.generate_seqs(num_seqs=5, seed=42)
cols = ['seq','op2.key.iupac_state','op1.key.positions','op1.key.wt_chars', 'op1.key.mut_chars']
df = df[cols]
df = df.rename(columns={col: col.split('.')[-1] for col in df.columns})
print(df)
```
